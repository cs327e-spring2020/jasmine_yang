## create modeled tables for jhu_csse

```python
!bq --location=US mk --dataset covid19_jhu_csse_modeled
```

```python
%%bigquery
create or replace table covid19_jhu_csse_modeled.summary
as select null as id, * from covid19_jhu_csse.summary
```

```python
%%bigquery
update covid19_jhu_csse_modeled.summary set id = FARM_FINGERPRINT(country_region) 
where province_state is null
```

```python
%%bigquery
update covid19_jhu_csse_modeled.summary set id = FARM_FINGERPRINT(concat(province_state,country_region)) 
where province_state is not null
```

```python
%%bigquery
select * from covid19_jhu_csse_modeled.summary
limit 12
```

```python
%%bigquery
select count(*) as null_id_count
from covid19_jhu_csse_modeled.summary
where id is null
```

```python
%%bigquery
select id, province_state, country_region
from covid19_jhu_csse_modeled.summary
order by province_state, country_region
limit 5
```

```python
%%bigquery
create or replace table covid19_jhu_csse_modeled.location as
select distinct id, province_state, country_region, latitude, longitude, fips, admin2, combined_key
from covid19_jhu_csse_modeled.summary
```

```python
%%bigquery
select * from covid19_jhu_csse_modeled.location
limit 12
```

```python
%%bigquery
select count(*) as location_count
from covid19_jhu_csse_modeled.location
```

get US location

```python
%%bigquery
create or replace table covid19_jhu_csse_modeled.us_location as
select distinct id, province_state, country_region, latitude, longitude, fips, admin2, combined_key
from covid19_jhu_csse_modeled.location
where country_region = 'US'
```

```python
%%bigquery
select * from covid19_jhu_csse_modeled.us_location
limit 5
```

```python
%%bigquery
select count(*) as us_location_count
from covid19_jhu_csse_modeled.us_location
```

```python
%%bigquery
create or replace table covid19_jhu_csse_modeled.reports as
select id as location_id, date, confirmed, deaths, recovered, active
from covid19_jhu_csse_modeled.summary
```

```python
%%bigquery
select count(*) as event_count
from covid19_jhu_csse_modeled.reports
```

```python
%%bigquery
select * from covid19_jhu_csse_modeled.reports
order by date
limit 5
```

```python
%%bigquery
select * from covid19_jhu_csse_modeled.reports
order by date desc
limit 5
```
