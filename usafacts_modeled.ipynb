## create modeled tables for usafacts

```python
!bq --location=US mk --dataset covid19_usafacts_modeled
```

```python
%%bigquery
create or replace table covid19_usafacts_modeled.loc_county as
select distinct cast(county_fips_code as int64) as county_fips_code, county_name
from covid19_usafacts.summary
```

```python
%%bigquery
select * from covid19_usafacts_modeled.loc_county
limit 12
```

```python
%%bigquery
create or replace table covid19_usafacts_modeled.loc_state as
select distinct cast(state_fips_code as int64) as state_fips_code, state
from covid19_usafacts.summary
```

```python
%%bigquery
select * from covid19_usafacts_modeled.loc_state
limit 12
```

```python
%%bigquery
create or replace table covid19_usafacts_modeled.reports as
select cast(county_fips_code as int64) as county_fips_code, cast(state_fips_code as int64) as state_fips_code, date, confirmed_cases, deaths
from covid19_usafacts.summary
```

```python
%%bigquery
select * from covid19_usafacts_modeled.reports
limit 12
```
